## Best model for the QSAR dataset

So, by looking at the analysis that is done above, we can say that the best overall model for this dataset is the Logistic Regression model where it has the highest accuracy compared to the rest of the model and it has only 20 features being used to do so. With that, our team concludes that the best overall machine learning model is the Logistic Regression model with 20 features. The 20 features are obtained through Backward SFS. 

Although Perceptron has similar result as the Logistic Regression model, it is not picked since it has more features used. 
That Perceptron model uses 40 features which is almost all of the features used in the dataset. Thus, we picked the Logistic 
Regression since it uses 20 features to produce that result.

### Creating back the Best Model down below
The best model is created, which is Logistic Regression model with 20 features that is obtained through Backward SFS.

The parameters for the LR model : C=10.0, penalty='l2'

```python
# getting the dataset
qsar = pd.read_csv('biodeg.csv', sep=';')
qsar = qsar.drop("Unnamed: 42", axis=1)  # removing the unnamed column if you have, no means, comment it out
```

```python
# changing the str label into 1s and 0s
qsar['experimental_class'] = qsar['experimental_class'].astype('category')
encode_map = {
    'RB': 1,
    'NRB': 0
}
qsar['experimental_class'].replace(encode_map, inplace=True)
# 1 reprsents RB and O represents NRB [changed it into a binary classification problem now]
```

```python
y = qsar['experimental_class']
cols = ['SpMax_L', 'J_Dz(e)', 'nHM', 'NssssC', 'nCb-', 'SdssC', 'HyWi_B(m)','LOC', 'SM6_L', 'Mi', 
        'nArNO2', 'Psi_i_1d', 'B04[C-Br]', 'SdO', 'nCrt','F02[C-N]', 'nHDon', 'Psi_i_A', 'nN', 'nArCOOR']
x = qsar[cols]
```

```python
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
```

```python
seed_num = 80
x_sfs, x_not_sfs, y_sfs, y_not_sfs = train_test_split(x_scaler, y, test_size=0.3, random_state=seed_num)
```

```python
x_test_ = x_not_sfs
y_train = y_sfs
y_test = y_not_sfs
```

```python
m_lr_best = LogisticRegression(C=10.0, penalty='l2')
m_lr_best.fit(x_sfs, y_train)
```

```python
# result based on the testing set
y_pred = m_lr_best.predict(x_test_)
print("Mean absolute error : " + str(mean_absolute_error(y_test, y_pred)))
print("Mean squared error : " + str(mean_squared_error(y_test, y_pred)))
print("r2 score : " + str(r2_score(y_test, y_pred)))
print("The max error value : " + str(max_error(y_test, y_pred)) + "\n")
print("accuracy score: ", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
